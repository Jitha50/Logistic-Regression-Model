```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# we have to load the data
data = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
data.head()
```

```python
# basic data preprocessing process
data.info()
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
# we will split opur data in x and y values
X = data.iloc[:,[2,3]].values
y = data.iloc[:,4].values
```

```python
# data splitting using train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 0)
```

```python
 # We have standarize our data and we have to reduce its size from high values
# to the lower bvalues
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()

X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# we have to preeict our values
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# The metrices we used for evaluating our model.
from sklearn import metrics
# we have find the confusion matrix 
# The column will show the instances predicted for each label
# and the rows will show the actual number of instance for each label

print(metrics.confusion_matrix(y_test,y_pred, labels = [0,1]))
```
